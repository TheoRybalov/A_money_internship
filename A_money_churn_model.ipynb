```python
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop


### Метод поиска пар с наибольшей корреляцией

def get_top_abs_correlation(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels_to_drop).sort_values(ascending = False)
    return au_corr[0:n]

print("Top absolute Correlations TRAIN")
print(get_top_abs_correlation(train[numeric_columns], 50))
```

```python
### Удалим колонки, где корреляция > 0.7

def correlation(dataset, threshold):
    col_corr = set()
    # Оставляем только числовые столбцы
    
    numeric_dataset = dataset.select_dtypes(include=[np.number])
    corr_matrix = numeric_dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname]

correlation(train, 0.8)
```

```python
cat_features = ["payment_frequency", "status", "source", "settlement", "interface", "type","client_type", "gender", "have_extension"]
```

```python


from sklearn.model_selection import train_test_split
X = train.drop(columns=['churn'])
y = train['churn']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train.dtypes
```

```python
for col in cat_features:
    print(f"{col}: {X_train[col].dtype}")
    X_train[col] = X_train[col].astype(str)
    X_val[col] = X_val[col].astype(str)
    print(f"{col}: {X_train[col].dtype}")
```

```python
X_train.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
numeric_columns = X_train.loc[:, X_train.dtypes != np.object_].columns
scaler = StandardScaler()
X_train[numeric_columns] = scaler.fit_transform(X_train[numeric_columns])
X_val[numeric_columns] = scaler.transform(X_val[numeric_columns])
```

```python
X_train.dtypes
```

```python
class_counts = pd.Series(y_train).value_counts()
print(f'Распределение классов:\n{class_counts}')
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.array([0, 1]),
    y=y_train  # y_train — ваш таргет
)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
```

```python
import catboost as cb

# # Обучение модели
model = cb.CatBoostClassifier(
    iterations=3000,
    depth=7,
    learning_rate=0.03,
    loss_function='Logloss',
    l2_leaf_reg=20,
    random_strength=3,
    bagging_temperature=1.2,
    random_seed=42,
    grow_policy='Lossguide',
    eval_metric='F1',
    class_weights=class_weights_dict,
    cat_features=cat_features,
    verbose=1
)

```

```python
from sklearn.metrics import f1_score

model.fit(X_train, y_train, eval_set=(X_val, y_val), early_stopping_rounds=50)

print("Признаки по важности")
# Получаем важность признаков
feature_importance = model.get_feature_importance(prettified=True)

# Создаем DataFrame
df = pd.DataFrame(feature_importance)

# Сортируем по важности
df = df.sort_values(by="Importances", ascending=False)

# Выводим в консоль
print(df.to_string(index=False))


# Оценка модели
y_pred = model.predict(X_val)
f1 = f1_score(y_val, y_pred)
print(f'F1-score: {f1:.4f}')
```
